```python
query = '''
WITH FirstDepartment AS (
    SELECT 
        edh.employee_id,
        edh.department_id AS first_department_id,
        d.department_name AS first_department_name
    FROM 
        employee_department_history edh
    JOIN 
        departments d ON edh.department_id = d.department_id
    WHERE 
        (edh.employee_id, edh.start_date) IN (
            SELECT 
                employee_id, MIN(start_date)
            FROM 
                employee_department_history
            GROUP BY 
                employee_id
        )
),
CurrentDepartment AS (
    SELECT 
        edh.employee_id,
        edh.department_id AS current_department_id,
        d.department_name AS current_department_name
    FROM 
        employee_department_history edh
    JOIN 
        departments d ON edh.department_id = d.department_id
    WHERE 
        edh.end_date IS NULL
),
DepartmentCount AS (
    SELECT 
        employee_id,
        COUNT(DISTINCT department_id) AS department_count
    FROM 
        employee_department_history
    GROUP BY 
        employee_id
)
SELECT 
    e.employee_id,
    e.employee_name,
    fd.first_department_name,
    cd.current_department_name,
    dc.department_count
FROM 
    employees e
JOIN 
    FirstDepartment fd ON e.employee_id = fd.employee_id
JOIN 
    CurrentDepartment cd ON e.employee_id = cd.employee_id
JOIN 
    DepartmentCount dc ON e.employee_id = dc.employee_id
ORDER BY 
    e.employee_id;
'''
```
