Ao criar um modelo e verificar que o resultado está abaixo do esperado, podemos seguir com alguns passos:

- Análise de dados:
Realizar uma análise exploratória para entender melhor as variáveis e suas distribuições.
Identificar e tratar outliers (se houver) que possam estar influenciando o modelo de maneira negativa.
Plotar uma matriz de correlação para entender como as variáveis se relacionam.

- Multicolinearidade:
Verificar se há multicolinearidade entre as variáveis independentes e, se necessário, excluir ou combinar variáveis.

- Selecionar Variáveis:
Utilizar métodos para selecionar as variáveis mais relevantes (ex.: Feature Importance).

- Maior quantidade de dados:
Depedendo do tamanho do dataset, aumentar a quantidade de dados nos dados de treino deixaria o modelo mais robusto.

- Cross-Validation:
Para avaliar a performance do modelo e evitar overfitting. Vale a pena ressaltar que Cross-Validation é indicado em datasets menores por conta do poderio computacional necessário para rodar.

- Pipeline de Machine Learning:
Pipelines são construídos para manter organizado o pré-processamento de dados e o código de modelagem, basicamente.
Pode-se criar um pipeline para tratar os dados (valores nulos, valores categóricos...), definir o modelo e avaliar o modelo. Com isso, o pipeline deixa o código mais "limpo", pois agrupa etapas de um processo de criação de modelo.

    É possível criar pipelines com a biblioteca *sklearn*, por exemplo. (*from sklearn.pipeline import Pipeline*)
